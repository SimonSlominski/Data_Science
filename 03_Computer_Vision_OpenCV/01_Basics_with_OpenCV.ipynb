## Import Libraries

```python
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

## Loading image

```python
img = cv2.imread(filename='view.jpg')

# By default the img is reading as a numpy array
img
```

```python
#Show the image with matplotlib
plt.imshow(img)
plt.show()
```

```python
#Show the image with matplotlib with real colors

# load image using cv2....and do processing.
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# as opencv loads in BGR format by default, we want to show it in RGB.
plt.show()
```

```python
# Show the size of the image
img.shape

# (480, 586, 3)
# height = 480
# width = 586
# colors = 3 --> 3 means it's a color photo -> RGB
```

## Saving the image to a file

```python
cv2.imwrite(filename='view-copy.png', img=img)
```

## Loading image in grayscale

```python
img = cv2.imread(filename='view.jpg', flags=cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
plt.show()
```

```python
# Img in grayscale has no 3rd element. Depth = None
img.shape
```

## Image Addition

```python
img1 = cv2.imread(filename='bear.jpg')
img2 = cv2.imread(filename='view.jpg')
```

```python
# Important: images must be the same size
add_img = cv2.add(src1=img1, src2=img2)
add_img.shape
```

```python
plt.imshow(add_img)
plt.show()
```

## Cutting out parts of the image

```python
img = cv2.imread(filename='view.jpg')
```

```python
# Cutting left top square
left_top_square = img[:200, :200]
plt.imshow(left_top_square)
plt.show()
```
