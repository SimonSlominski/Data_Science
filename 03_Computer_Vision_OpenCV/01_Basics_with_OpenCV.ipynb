### Image resizing

```python
img = cv2.imread(filename = 'python.png')
plt.imshow(img)
plt.show()
```

```python
img_resized = imutils.resize(image=img, height=300)
plt.imshow(img_resized)
plt.show()
```

```python
img_resized = imutils.resize(image=img, width=300)
plt.imshow(img_resized)
plt.show()
```

### Rotation

```python
height, width = img.shape[:2]

M = cv2.getRotationMatrix2D(center=(height // 2, width // 2), angle=45, scale=0.5)
plt.imshow(cv2.warpAffine(src=img, M=M, dsize=(width, height)))
plt.show()
```

### Thresholding (image processing)

```python
img_gray = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.threshold(src=img_gray, thresh=50, maxval=255, type=cv2.THRESH_BINARY)[1])
plt.show()

# Values up to 50 will remain unchanged, above 50 will be converted to white
# 0 == black
# 255 == white
```

### Edge detection

```python
logo = cv2.imread('python.png')
```

```python
# Vertical edges
kernel_vertical = np.array([[-1, 0, 1],
                            [-2, 0, 2],
                            [-1, 0, 1]])

output = cv2.filter2D(src=logo, ddepth=-1, kernel=kernel_vertical)
plt.imshow(output)
plt.show()
```

```python
# Horizontal edges
kernel_horizontal = np.array([[-1, -2, -1],
                            [0, 0, 0],
                            [1, 2, 1]])

output = cv2.filter2D(src=logo, ddepth=-1, kernel=kernel_horizontal)
plt.imshow(output)
plt.show()
```

```python
plt.imshow(cv2.Canny(image=logo, threshold1=100, threshold2=100))
plt.show()
```

### Erode and dilate

```python
logo = cv2.imread('python.png')
```

```python
kernel = np.ones((5, 5), np.uint8)
```

```python
plt.imshow(cv2.erode(src=logo, kernel=kernel, iterations=1))
plt.show()
```

```python
plt.imshow(cv2.dilate(src=logo, kernel=kernel, iterations=1))
plt.show()
```

## Drawing

### Adding a rectangle

```python
img = cv2.imread('bear.jpg')
```

```python
# pt1: upper left corner, pt2: bottom right corner
plt.imshow(cv2.rectangle(img=img, pt1=(75, 80), pt2=(180, 200), color=(0, 0, 255), thickness=2))
plt.show()
```

```python
plt.imshow(cv2.rectangle(img=img, pt1=(140, 250), pt2=(350, 470), color=(0, 0, 255), thickness=2))
plt.show()
```

### Adding a circle

```python
img = cv2.imread('bear.jpg')
copy_img = img.copy()
```

```python
plt.imshow(cv2.circle(img=copy_img, center=(300, 300), radius=20, color=(255, 0, 0), thickness=3))
plt.show()
```

### Adding lines
